練習 2：數值最小值計算 Numerical minimization

計算函數 $f(x) = x^4 - 8x^3+16x^2-2x+8$ 的最小值。

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
 
# draw the graph
coef = [1, -8, 16, -2, 8] # high to low 
f= lambda x : np.polyval(coef, x)
x = np.linspace(-1, 5, 1000)
plt.plot(x, f(x))
plt.xlabel('X'), plt.grid(True)
 
# compute the local minimum
res = opt.minimize_scalar(f, bracket=[2, 3, 5]) #method='Brent')
print(res) # find out the return values
print('The function has a local minimum at x = {:.4f}'.format(res.x))
print('The corresponding function value is {:.4f}'.format(res.fun))
 
plt.text(res.x, res.fun, 'X', color = 'r',
    horizontalalignment='center',
    verticalalignment='center')
plt.title('The local minimum of $f(x)$')
plt.show()
```

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
 
# First, draw function graph
# define a function for a complicated funcion
def V(t) :
    tmp = ((10*np.pi - 5*t) / (2*np.pi))**2
    V = np.pi / 3 * tmp * np.sqrt(25 - tmp)
    return V
 
t = np.linspace(0, 3*np.pi, 1000)
plt.plot(t, V(t))
 
f = lambda x : -V(x)
res = opt.minimize_scalar(f, bounds = [0, 2])
 
plt.text(res.x, -res.fun, '+', color ='r', fontsize = 16,
    horizontalalignment='center',
    verticalalignment='center')
plt.xlabel('X'), plt.grid(True)
plt.title('The fnction maximum is {:.4f} at x = {:.4f}'.format(-res.fun, res.x))
plt.show()
```

### 習題2: $\displaystyle \min_{-4 \leq x \leq 3} (x+1)^5 sin(x-3)$

1. 計算在此區間函數的最小值

```python
import numpy as np
from scipy.optimize import minimize

# 定義目標函數
def target_function(x):
    return (x + 1)**5 * np.sin(x - 3)

# 隨機生成初始猜測值
initial_guess = np.random.uniform(-4, 3)

# 找到最小值
min_result = minimize(target_function, initial_guess)

# 最小值及對應的x值
min_value = min_result.fun
min_x = min_result.x

# 顯示結果
print(f"min: {min_value}")
print(f"x value: {min_x}")

```

2. 繪製函數圖形並標記最小值位置

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# 定義目標函數
def target_function(x):
    return (x + 1)**5 * np.sin(x - 3)

# 指定範圍
bounds = (-4, 3)

# 生成x值範圍
x_values = np.linspace(bounds[0], bounds[1], 10000)

# 計算對應的y值
y_values = target_function(x_values)

# 找到最小值
min_result = minimize_scalar(target_function, bounds=bounds, method='bounded')

# 最小值及對應的x值
min_value = min_result.fun
min_x = min_result.x

# 繪製圖形
plt.plot(x_values, y_values, label='Target Function', color='blue')
plt.scatter(min_x, min_value, color='red', marker='x', label='Min Value')
plt.xlabel('x')
plt.ylabel('y')
plt.title('$\min_{-4 \leq x \leq 3} (x+1)^5 sin(x-3)$')
plt.annotate(f'Min Value: x={min_x:.2f}, y={min_value:.2f}', (min_x, min_value),
            textcoords="offset points", xytext=(-30, 0), ha='right', fontsize=10,color='red')
plt.legend()
plt.grid(True)
plt.show()
```

3. 找出實數根位置並標記在函數圖形上

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar, fsolve

# 定義目標函數
def target_function(x):
    return (x + 1)**5 * np.sin(x - 3)

# 指定範圍
bounds = (-4, 3)

# 生成x值範圍
x_values = np.linspace(bounds[0], bounds[1], 10000)

# 計算對應的y值
y_values = target_function(x_values)

# 找到最小值
min_result = minimize_scalar(target_function, bounds=bounds, method='bounded')

# 最小值及對應的x值
min_value = min_result.fun
min_x = min_result.x

# 尋找所有實數根
roots_values = fsolve(target_function, x0=np.linspace(bounds[0], bounds[1], 5))

# 繪製圖形
plt.plot(x_values, y_values, label='Target Function', color='blue')
plt.scatter(min_x, min_value, color='red', marker='x', label='Min Value')
plt.scatter(roots_values, [0]*len(roots_values), color='green', marker='o', label='Real Roots', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$\min_{-4 \leq x \leq 3} (x+1)^5 \sin(x-3)$')
plt.annotate(f'Min Value: x={min_x:.2f}, y={min_value:.2f}', (min_x, min_value),
            textcoords="offset points", xytext=(-30, 0), ha='right', fontsize=10, color='red')
plt.legend()
plt.grid(True)
plt.show()
print(f"Real Roots: {roots_values}")

```

### 習題3: 計算 L(x) = 10 的解 x, 其中
$L(x) = \int_a^x \sqrt{1 + (f'(t))^2}\; dt, \;\; for f(t) = t^2/2 and a = 0.$


1. 計算 L(x) = 10 的解 x

```python
import numpy as np
from scipy.integrate import quad

# 定義函數 f(t) 和其導數 f'(t)
def f(t):
    return t**2 / 2

def df_dt(t):
    return t

# 定義積分的被積函數
def integrand(t):
    return np.sqrt(1 + df_dt(t)**2)

# 指定目標積分值
target_L = 10

# 計算積分
result, _ = quad(integrand, 0, np.inf)

# 定義解方程的函數
def equation_to_solve(x):
    result, _ = quad(integrand, 0, x)
    return result - target_L

# 使用fsolve解方程，找到對應的x
from scipy.optimize import fsolve
initial_guess = 1  # 可以根據問題特性調整初始猜測值
solution = fsolve(equation_to_solve, initial_guess)[0]

print(f"The solution for L(x) = 10 is x = {solution}")
```

2. 繪製函數圖形並標記解 x 的位置

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.optimize import minimize_scalar

# 定義函數 f(t) 和其導數 f'(t)
def f(t):
    return t**2 / 2

def df_dt(t):
    return t

# 定義積分的被積函數
def integrand(t):
    return np.sqrt(1 + df_dt(t)**2)

# 定義目標函數 L(x)
def target_function(x):
    result, _ = quad(integrand, 0, x)
    return result

# 找到最小值（最小值即為 L(x) = 10 的解）
min_result = minimize_scalar(lambda x: np.abs(target_function(x) - 10), bounds=[0, 10], method='bounded')
min_x = min_result.x
min_value = target_function(min_x)

# 生成 x 值範圍
x_values = np.linspace(0, 10, 400)
y_values = [target_function(x) for x in x_values]

# 繪製目標函數圖形
plt.plot(x_values, y_values, label=r'$L(x) = \int_{0}^{x} \sqrt{1 + (f\'(t))^2}\; dt$',color='blue')
plt.scatter(min_x, min_value, marker='x', label='L(x)=10',color='red')
plt.annotate(f'L(x)=10: x={min_x:.2f}', (min_x, min_value),
            textcoords="offset points", xytext=(60, 0), ha='center', fontsize=10,color='red')
plt.xlabel('x')
plt.ylabel('L(x)')
plt.title('$L(x) = \int_{0}^{x} \sqrt{1 + (f\'(t))^2}\; dt$')
plt.legend()
plt.grid(True)
plt.show()
```

**<font color=skyblue>注意事項與討論：</font>**
- f(t) 和 df_dt(t) 定義了原函數以及其導數
- integrand(t) 定義了積分的被積函數
- target_function(x)給定函數的弧長積分
- 使用 fsolve 函數尋找目標函數的實數根

### 習題4: 最大概似函數估計（MLE）（離散分配）：計算 $\displaystyle \max_{p} \ln \Pi_{i=1}^N f(x_i; p) $,
其中現自幾何分配 (p=0.2) 抽取樣本 x_i, i=1,2,\cdots, N，且令樣本數分別為 N= 10, 20 ,30, 50, 100, 300, 500,1000，欲採最大概似估計法(log MLE)估計 p。

1. 任取一組樣本，繪製目標函數 $\displaystyle \ln \Pi_{i=1}^N f(x_i; p)$，並標示出最大值的位置。

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# 設定真實的機率值
true_p = 0.2
# 設定樣本數
N = [10]

# 產生一個均勻分布的機率值範圍
p_range = np.linspace(0, 1, 10000)

# 創建一個畫布和子圖
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

# 生成符合幾何分布的隨機樣本
s = np.random.geometric(true_p, size=N[0])

# 定義似然函數
log_Likelihood = lambda x: -(N[0] * np.log(x) + np.log(1 - x) * (np.sum(s) - N[0]))

# 進行最大概似估計
res = opt.minimize_scalar(log_Likelihood, bounds=[0, 1], method='bounded')

# 繪製似然函數曲線
ax.plot(p_range, -log_Likelihood(p_range),color='blue')
# 標記最大似然估計的位置
ax.plot(res.x, -res.fun, 'ro')
# 在圖上標示最大似然估計的值
ax.annotate(f'Max $p$: {res.x:.2f}', (res.x, -res.fun),
            textcoords="offset points", xytext=(50, -20), ha='center', fontsize=12,color='red')

# 加入網格線
ax.grid(True, linewidth=0.5)

# 在整張圖下方添加 x 軸標籤
fig.text(0.5, 0.05, r'$\hat{p}$', ha='center', va='center')

# 整個圖的標題
fig.suptitle('$\max_{p} \; \ln\prod_{i=1}^N (1-p)^{x_i-1}p\;\;,p=0.2\;\;,N=10$')

# 顯示圖形
plt.show()


```

2. 繪製樣本數 N 與估計值 p 的關係圖。

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# 設定真實的機率值
true_p = 0.2
# 設定多個樣本數
N_values = [10, 20, 30, 50, 100, 300, 500, 1000]

# 產生一個均勻分布的機率值範圍
p_range = np.linspace(0, 1, 10000)

# 創建一個畫布和子圖
fig, ax = plt.subplots(2, 4, figsize=(20, 10))
ax = ax.flatten()  # 展平子圖陣列，以便適應不同的樣本數

for i, N in enumerate(N_values):
    # 生成符合幾何分布的隨機樣本
    s = np.random.geometric(true_p, size=N)

    # 定義似然函數
    log_Likelihood = lambda x: -(N * np.log(x) + np.log(1 - x) * (np.sum(s) - N))

    # 進行最大概似估計
    res = opt.minimize_scalar(log_Likelihood, bounds=[0, 1], method='bounded')

    # 繪製似然函數曲線
    ax[i].plot(p_range, -log_Likelihood(p_range),color='blue')
    # 標記最大似然估計的位置
    ax[i].plot(res.x, -res.fun, 'ro')
    # 在圖上標示最大似然估計的值
    ax[i].annotate(f'Max $p$= {res.x:.2f}', (res.x, -res.fun),
                   textcoords="offset points", xytext=(50, -20), ha='center', fontsize=12,color='red')

    # 加入網格線
    ax[i].grid(True, linewidth=0.5)

    # 設定子圖標題
    ax[i].set_title(r'$N = %d$' % N, fontsize=12)

# 在整張圖下方添加 x 軸標籤
fig.text(0.5, 0.05, r'$\hat{p}$', ha='center', va='center')

# 整個圖的標題
fig.suptitle('$\max_{p} \; \ln\prod_{i=1}^N (1-p)^{x_i-1}p\;\;,p=0.2$', fontsize=18)

# 調整子圖之間的間距
plt.tight_layout()

# 顯示圖形
plt.show()
```

3. 繪製折線圖，呈現每組樣本數的 p 估計值和標準差。

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from scipy.stats import geom  # 使用幾何分佈

M = 10000
true_p = 0.2  # 幾何分佈的成功概率
N = [10, 20, 30, 50, 100, 200, 300, 500, 1000]
Mu = np.zeros(len(N))
Std = np.zeros(len(N))

for j in range(len(N)):
    S = geom.rvs(p=true_p, size=(M, N[j]))
    R = np.zeros(M)
    for i in range(M):
        f = lambda l: -(N[j] * np.log(l) - l * np.sum(S[i, :]))
        res = opt.minimize_scalar(f, bounds=[0.01, 5], method='bounded')
        R[i] = res.x
    Mu[j], Std[j] = R.mean(), R.std()

print("Mean of the Posterior Distribution:")
print(Mu)
print("\nStandard Deviation of the Posterior Distribution:")
print(Std)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(Mu, marker='s', color='b')
ax1.set_xticks(np.arange(len(N)))
ax1.set_xticklabels(N)
ax1.grid(which='minor')
ax1.set_xlabel('Sample Size')
ax1.set_title('Mean of the Posterior Distribution')

ax2.plot(Std, marker='s', color='r')
ax2.set_xticks(np.arange(len(N)))
ax2.set_xticklabels(N)
ax2.grid(which='minor')
ax2.set_xlabel('Sample Size')
ax2.set_title('Standard Deviation of the Posterior Distribution')

plt.show()
```
