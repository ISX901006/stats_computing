$$f(x) = \frac{\sin(x)}{x}, \;-3\pi \leq x \leq 3\pi ，並從圖猜測 \lim_{x\rightarrow 0} f(x) = ? \\再加入第二與第三個函數 \displaystyle  g(x) = \frac{\sin(x^2)}{x}, h(x) = \frac{\sin^2(2x)}{x^2}，同時猜測 \lim_{x\rightarrow 0} g(x) = ?,  \lim_{x\rightarrow 0} h(x) = ?$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Function f(x) = sin(x)/x
x = np.linspace(-3*np.pi, 3*np.pi, 1000)  # Generate values for x
y_f = np.where(x == 0, 1, np.sin(x) / x)  # Handle the special case when x = 0

# Function g(x) = sin(x^2)/x
y_g = np.where(x == 0, 0, np.sin(x**2) / x)

# Function h(x) = sin^2(2x)/x^2
y_h = np.where(x == 0, 0, (np.sin(2*x)**2) / (x**2))

# Create a figure with three subplots
plt.figure(figsize=(12, 4))
plt.subplot(131)  # First subplot
plt.plot(x, y_f, color='red', label='f(x)')  # Plot the graph of f(x)
plt.title('f(x) = sin(x)/x', pad=10, fontweight='bold')  # Add a subplot title
plt.ylim(-0.5, 1.5)  # Set the y-axis range
plt.xlabel('x')  # Set the x-axis label
plt.ylabel('f(x)')  # Set the y-axis label
plt.grid(True)  # Add gridlines
plt.legend()

plt.subplot(132)  # Second subplot
plt.plot(x, y_g, color='orange', label='g(x)')  # Plot the graph of g(x)
plt.title('g(x) = sin(x^2)/x', pad=10, fontweight='bold')  # Add a subplot title
plt.ylim(-1, 1)  # Set the y-axis range
plt.xlabel('x')  # Set the x-axis label
plt.ylabel('g(x)')  # Set the y-axis label
plt.grid(True)  # Add gridlines
plt.legend()

plt.subplot(133)  # Third subplot
plt.plot(x, y_h, color='green', label='h(x)')  # Plot the graph of h(x)
plt.title('h(x) = sin^2(2x)/x^2', pad=10, fontweight='bold')  # Add a subplot title
plt.ylim(-0.5, 4.5)  # Set the y-axis range
plt.xlabel('x')  # Set the x-axis label
plt.ylabel('h(x)')  # Set the y-axis label
plt.grid(True)  # Add gridlines
plt.legend()

# Display the plots
plt.tight_layout()  # Automatically adjust subplot layout to prevent overlap
plt.show()
```

**<font color=skyblue>注意事項與討論：</font>**
- 引入NumPy和Matplotlib庫，用於數值計算和繪圖。
- 以足夠多的項次，代替無限次項的加法。
- 調整Y軸的範圍以達到較佳的觀賞角度，X軸則固定。
- x 數組將包含1000個數值，在從 -3π 到 3π 的範圍內均勻分佈，這些數值可以用作繪製函數的 x 座標或用於進行各種數學計算的數據。
- 牽涉三角函數，因此 X 軸範圍以 $\pi$ 的倍數訂定。
- 將x軸的0置中，方便觀察函數其x值最接近0時之y值。

> 根據觀察，有以下結論:<br>
  >    - 當 x 趨近於 0 時，f(x) 的極限等於 1。 <br>
  >    - 當 x 趨近於 0 時，g(x) 的極限等於 0。 <br>
  >    - 當 x 趨近於 0 時，h(x) 的極限等於 4。 <br>

$$f(x) = \frac{e^{\alpha x}}{e^{\alpha x}+1},\alpha=1\\
畫出漸進線和函數爬升的趨勢$$ 



```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return np.exp(x) / (np.exp(x) + 1)

# Define the range of x values
x = np.linspace(-8, 8, 800)

# Calculate the corresponding y values
y = f(x)

# Plot the function curve
plt.plot(x, y, label='f(x) = e^(αx) / (e^(αx) + 1)', color='blue')

# Plot the asymptotic line
asymptotic_line = np.ones_like(x)
plt.plot(x, asymptotic_line, label='Asymptotic Line (y = 1)', linestyle='--', color='red')

# Set labels for x and y axes
plt.xlabel('x')
plt.ylabel('f(x)')

# Set the title of the plot
plt.title('Sigmoid function', pad=10, fontweight='bold')

# Add a legend to the plot
plt.legend()

# Display a grid on the plot
plt.grid(True)

arrow_x = 0  # x-coordinate of the arrow
arrow_y = 0.5  # y-coordinate of the arrow
arrow_dx = 2  # change in x for the arrow
arrow_dy = 0.2  # change in y for the arrow
plt.annotate('Climb', xy=(arrow_x, arrow_y), xytext=(arrow_x + arrow_dx, arrow_y + arrow_dy),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12, color='green')
# Show the plot
plt.show()
```

**<font color=skyblue>注意事項與討論：</font>**
- 引入NumPy和Matplotlib庫，用於數值計算和繪圖。
- 創建 x 值範圍，它生成一組 x 值，這些值範圍從 -8 到 8，共有 800 個點，可以清楚看到函數漸進和上升的趨勢。
- 計算 y 值：它使用函數 f(x) 計算對應的 y 值，這些 y 值將在 y 軸上繪製函數的曲線。
- 使用 Matplotlib 繪製函數曲線，並使用plt.legend()標記為 "f(x) = e^(αx) / (e^(αx) + 1)"，以及使用藍色表示。
- 紅色虛線 y = 1，表示函數的漸近線，並同樣以plt.legend()標記圖例。
- plt.x/y label設置 x 和 y 軸的標籤。
- plt.grid()在圖形上顯示網格線，以便識別函數曲線和漸近線。
- 使用 plt.annotate() 函數在圖形上添加一個箭頭註釋，以指示函數的爬升部分。

S型函數是一種函數，因其函數圖像形狀像字母S得名。其形狀曲線至少有2個焦點，也叫「二焦點曲線函數」。S型函數是有界、可微的實函數，在實數範圍內均有取值，且導數恆為非負，有且只有一個拐點。

$$f(x) = {e^{-x/10}}\sin(x)\\畫出震盪幅度逐漸變小的趨勢$$ 



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return np.exp(-x / 10) * np.sin(x)

# Define the range of x values
x = np.linspace(0, 70, 1000)

# Calculate the corresponding y values
y = f(x)

# Find the indices of the maximum and minimum points
max_index = np.argmax(y)
min_index = np.argmin(y)

# Plot the function curve
plt.plot(x, y, label='$f(x) = e^{-x/10} \sin(x)$', color='blue')

# Annotate the values of the maximum and minimum points
plt.annotate(f'Max: ({x[max_index]:.1f}, {y[max_index]:.1f})', 
             xy=(x[max_index], y[max_index]), 
             xytext=(x[max_index] -12, y[max_index] +0.3), 
             arrowprops={'arrowstyle': '->', 'color': 'red'})
plt.annotate(f'Min: ({x[min_index]:.1f}, {y[min_index]:.1f})', 
             xy=(x[min_index], y[min_index]), 
             xytext=(x[min_index] -12, y[min_index] -0.3), 
             arrowprops={'arrowstyle': '->', 'color': 'green'})

# Set labels for x and y axes
plt.xlabel('x')
plt.ylabel('f(x)')

# Set the title of the plot
plt.title('Decreasing Oscillation Amplitude', fontweight='bold', pad=20)

# Add a legend to the plot
plt.legend()

# Display a grid on the plot
plt.grid(True)
# Show the plot
plt.show()

```

- 定義 x 的範圍從0到70，並且在這個範圍內生成了1000個數值點，以用於繪圖。
- f(x) 計算了每個 x 值對應的 y 值，這些 y 值將用於繪製函數的曲線。
- 使用 NumPy 函數找到函數曲線上的最高點和最低點的索引，並使用plt.annotate()添加了文字標籤，標記出最高點和最低點的數值，同時添加了箭頭以指示標籤位置。
- 使用 Matplotlib 繪製了函數的曲線，同時標記它為 "$f(x) = e^{-x/10} \sin(x)$"，並以藍色表示。
 ：我們設置了 x 軸和 y 軸的標籤，以說明圖形中的數據意義。

$$f(x) = \frac{1}{x-2}\\畫出漸近線的趨勢。分母為 0 的計算會引發警告，可以在設定 x 軸範圍時避開。\\譬如， np.setdiff1d(np.linspace(-5, 5, 200),[0])$$ 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of x values, excluding x = 2
x = np.setdiff1d(np.linspace(-5, 9, 200), [2])

# Calculate corresponding y values
y = 1 / (x - 2)

# Plot the function curve
plt.plot(x, y, label='f(x) = 1/(x-2)')

# Plot a vertical asymptote line at x = 2
plt.axvline(x=2, color='r', linestyle='--', label='Vertical Asymptote (x=2)')

# Set plot attributes
plt.ylim(-10, 10)  # Set the y-axis limits
plt.xlabel('x')    # Label the x-axis
plt.ylabel('f(x)')  # Label the y-axis
plt.title('f(x) = 1/(x-2)',fontweight='bold', pad=10)  # Set the plot title
plt.legend()  # Display the legend
# Display the plot
plt.grid(True)  # Add a grid to the plot
plt.show()


```

**<font color=skyblue>注意事項與討論：</font>**
- 引入NumPy和Matplotlib庫，用於數值計算和繪圖。
- 定義 x 的範圍從-5到9之間，並且在這個範圍內生成200個數值點，以用於繪圖。
- f(x) 計算了每個 x 值對應的 y 值，這些 y 值將用於繪製函數的曲線。
- 使用np.setdiff1d函數排除x等於2的值，因為函數在x=2處有一個垂直漸近線，所以不能在該點繪製圖形。
- 使用plt.plot函數，將x和y值繪製成函數的曲線，同時指定了曲線的標籤為"f(x) = 1/(x-2)"。 
- 使用plt.axvline函數，在x=2的位置繪製一條紅色虛線，表示垂直漸近線，同時指定該虛線的標籤為"Vertical Asymptote (x=2)"。
- 設置圖形的標題和設置 x 和 y 軸的標籤，並加粗顯示，以凸顯圖形的主題。
- plt.ylim(-10, 10)設定y軸的範圍
- 添加了圖例，以區分函數曲線和其他元素，並用plt.grid(True)在圖形上顯示了網格線，方便觀看函數趨勢。
- 使用plt.show()顯示了繪製好的圖形。

這個函數的定義範圍（domain）為除了 x = 2 以外的所有實數，因為當 x = 2 時分母變為零，導致函數未定義。這個函數的圖形通常是一條垂直於 x 軸的直線，並且在 x = 2 這個點有一個漸近線。當 x 靠近 2 時，f(x) 的絕對值會趨近無限大正或負，因此函數在 x = 2 處有一個垂直測地線的漸近行為。

$$f(x) = x^3+2\\繪製 f(x) 與 f^{-1}(x) 在同一張圖上並展示其互為反函數的對稱性。$$ 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of x values
x = np.linspace(-10, 10, 400)  # We choose this range because it covers the function's domain (x >= 2).

# Calculate the y values for f^{-1}(x)
f_inverse_x = np.cbrt(x - 2)  # Using the numpy cbrt function to calculate the cube root, f^{-1}(x) is this function

# Calculate the y values for f(x)
f_x = x**3 + 2  # We changed the original g(x) to f(x) to rename it as f(x)

# Plot the curve for f^{-1}(x)
plt.plot(x, f_inverse_x, label='$f^{-1}(x) = \\sqrt[3]{x - 2}$', color='b')  # Blue color represents the curve of f^{-1}(x)
# Plot the curve for f(x)
plt.plot(x, f_x, label='$f(x) = x^3 + 2$', color='r')  # Red color represents the curve of f(x)

# Plot the diagonal line y=x
plt.plot(x, x, color='y', linestyle='--', label='y=x')  # Yellow dashed line represents the diagonal line y=x

# Set graph attributes
plt.ylim(-10, 10)  # Set the y-axis range
plt.xlabel('x')  # Label for the x-axis
plt.ylabel('y')  # Label for the y-axis
plt.title('$f(x) & f^{-1}(x)$', fontweight='bold', pad=10)  # Title of the graph (bold)
plt.legend()  # Show the legend for labels

# Display the graph
plt.grid(True)  # Display gridlines
plt.show()
```

**<font color=skyblue>注意事項與討論：</font>**
- 引入NumPy和Matplotlib庫，用於數值計算和繪圖。
- 定義 x 的範圍從-10到10之間，並且在這個範圍內生成200個數值點，選擇這個範圍是因為函數的定義範圍要大於等於2。
- 計算函數f(x)反函數的對應y值，使用了numpy的cbrt函數計算了x−2的立方根，並將其保存在f_inverse_x中。
- 計算函數f(x)的對應y值，x的立方根再加上2，並將其保存在f_x中。
- 使用plt.plot函數，繪製了以上兩個函數和對角線，並同時嵌入標籤和設定顏色。
- 設定圖形的屬性，包括y軸範圍、x軸和y軸的標籤、圖形的標題以及圖例。
- 使用plt.grid(True)添加了網格線，並使用plt.show()顯示了繪製好的圖形。

這段程式碼用於視覺化兩個數學函數$f(x) = x^3 + 2$與$f^{-1}(x) = \sqrt[3]{x - 2}$ ，並添加了一條對角線y=x 以強調它們之間的對稱性。

$$f(x) = \frac{1}{\sqrt{2π}}e^{-{\frac{(x-3)^2}{2}}}\\畫出這張知名圖形的特色$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a range of x values
x = np.linspace(-4, 10, 400)

# Calculate the corresponding y values (probability density function of the normal distribution)
y = norm.pdf(x, loc=3, scale=1)  # Mean (loc) is 3, Standard Deviation (scale) is 1

# Plot the function curve
plt.plot(x, y, label='f(x)', color='black', linewidth='0.7')

# Set graph attributes
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Normal Distribution', pad=10, fontweight='bold')
plt.legend()

# Display gridlines
plt.grid(True)

# Calculate the area under the normal distribution curve

# Z-score range for the 68% area
z68_left = norm.ppf(0.16, loc=3, scale=1)  # 0.16 corresponds to the left 16% cumulative probability
z68_right = norm.ppf(0.84, loc=3, scale=1)  # 0.84 corresponds to the right 84% cumulative probability

# Z-score range for the 95% area
z95_left = norm.ppf(0.025, loc=3, scale=1)  # 0.025 corresponds to the left 2.5% cumulative probability
z95_right = norm.ppf(0.975, loc=3, scale=1)  # 0.975 corresponds to the right 97.5% cumulative probability

# Z-score range for the 99.7% area
z997_left = norm.ppf(0.0015, loc=3, scale=1)  # 0.0015 corresponds to the left 0.15% cumulative probability
z997_right = norm.ppf(0.9985, loc=3, scale=1)  # 0.9985 corresponds to the right 99.85% cumulative probability

# Plot the 68% area
plt.fill_between(x, 0, y, where=(x >= z68_left) & (x <= z68_right), color='red', alpha=0.5, label='68% Area')

# Plot the 95% area
plt.fill_between(x, 0, y, where=((x >= z95_left) & (x <= z95_right)) & ~((x >= z68_left) & (x <= z68_right)), color='orange', alpha=0.5, label='95% Area')

# Plot the 99.7% area
plt.fill_between(x, 0, y, where=(x >= z997_left) & (x <= z997_right) & ~((x >= z95_left) & (x <= z95_right)) & ~((x >= z68_left) & (x <= z68_right)), color='green', alpha=0.5, label='99.7% Area')

# Display the legend
plt.legend()

# Display the graph
plt.show()




```

**<font color=skyblue>注意事項與討論：</font>**
- 導入了必要的Python庫，包括numpy用於數值計算，matplotlib.pyplot用於繪圖，以及scipy.stats.norm用於處理正態分佈。
- 生成一個x值的範圍，這個範圍在-4到10之間，並總共包含了400個均勻分佈的點。
- 使用norm.pdf函數計算了正態分佈的概率密度函數，並將計算結果存儲在變數y中。這裡使用的正態分佈具有均值（loc）為3，標準差（scale）為1。
- 使用plt.plot函數繪製了正態分佈的概率密度函數曲線，並標記為'f(x)'，顏色為黑色，線寬度為0.7。
- 計算正態分佈曲線下的不同信心水平的區域，也就是68%、95%和99.7%信心區間的Z分數範圍，分別以紅色、黃色和綠色呈現。
- 使用plt.fill_between函數將不同信心區間的區域以不同的顏色填充，突顯了這些區域在正態分佈曲線下的位置。
- 設定圖形的屬性，包括x軸和y軸的標籤、圖形的標題，以及圖例，使用plt.grid(True)顯示了圖形的網格線。

常態分布，又名高斯分布、正規分布，是一個非常常見的連續機率分布。常態分布在統計學上十分重要，經常用在自然和社會科學來代表一個不明的隨機變量。

$$f(x) = 3x^3-x^4\\透過 X 軸範圍的選擇，觀察函數是否有最大值？是否看得出函數有幾個實數根？$$ 


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3*x**3 - x**4

# Generate a range of x values
x = np.linspace(-2, 5, 400)

# Calculate the corresponding y values
y = 3*x**3 - x**4

# Plot the function graph
plt.plot(x, y, label='f(x) = 3x^3 - x^4')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = 3x^3 - x^4', pad=10, fontweight='bold')
plt.grid(True)

# Find the maximum value and real roots
max_x = x[np.argmax(y)]  # x-coordinate of the maximum value
max_y = max(y)           # y-coordinate of the maximum value
roots = x[np.where(np.isclose(y, 0))]  # x-coordinates of real roots

# Mark the maximum value and real roots
plt.scatter(max_x, max_y, color='red', label='Maximum', zorder=5)
plt.scatter(roots, np.zeros_like(roots), color='green', label='Real Roots', zorder=5)

# Add coordinate labels
for root in roots:
    plt.text(root-0.07, 0.2, f'({root:.0f}, 0)', ha='right', va='bottom')

plt.text(max_x, max_y+0.3, f'({max_x:.2f}, {max_y:.2f})', ha='left', va='bottom')

# Adjust the y-axis range
plt.ylim(-2, 10) 

# Display the graph
plt.legend(loc='upper left')
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of x values, avoiding x less than or equal to zero
x = np.linspace(0.01, 15, 1000)

# Calculate the corresponding y values
y = np.log(x) / (x**2)

# Plot the function curve
plt.plot(x, y, label='f(x) = ln(x) / x^2', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = ln(x) / x^2', pad=10, fontweight='bold')

# Find the maximum value point
max_x = x[np.argmax(y)]
max_y = max(y)

# Plot the maximum value point
plt.scatter(max_x, max_y, color='red', marker='o', label=f'Max Value')

# Add a label for the maximum value
plt.text(max_x + 0.3, max_y - 0.005, f'({max_x:.2f}, {max_y:.2f})', ha='left', va='bottom')

# Plot the right asymptote (as x approaches infinity)
plt.axhline(0, color='green', linestyle='--', label='Asymptote (x -> ∞)')

# Mention that line tends to flatten when x increases
plt.annotate('Tends to Flatten', xy=(9, 0.03), xytext=(7, -0.075),
             arrowprops=dict(arrowstyle='->', color='black'), fontsize=12)
# Set graph attributes
plt.legend()
plt.grid(True)
plt.ylim(-0.2, 0.2)  # Set the y-axis range to clearly display the right asymptote

# Display the graph
plt.show()
```

$$f(x) = x^2 + 2x + 1 \\ 梯度下降$$ 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define an example function, in this case, a quadratic function
def sample_function(x):
    return x**2 + 2*x + 1

# Define the gradient descent algorithm
def gradient_descent(learning_rate, num_iterations):
    x = -5  # Initial value of x
    history = []  # Used to store the values at each iteration
    
    for _ in range(num_iterations):
        gradient = 2*x + 2  # Derivative of the function
        x -= learning_rate * gradient  # Update the value of x
        history.append(x)
    
    return history

# Set the learning rate and the number of iterations
learning_rate = 0.1
num_iterations = 50

# Optimize the function using gradient descent
trajectory = gradient_descent(learning_rate, num_iterations)

# Plot the function and the gradient descent process
x_values = np.linspace(-6, 1, 1000)
y_values = sample_function(x_values)
plt.plot(x_values, y_values, label='x^2 + 2x + 1')
plt.scatter(trajectory, [sample_function(x) for x in trajectory], color='red', label='Gradient Descent')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Optimization',pad=10)
plt.show()

```

**<font color=skyblue>注意事項與討論：</font>**
- 這個函數是一個二次函數，表示為 f(x) = x^2 + 2x + 1。這是一個凹向上的曲線，將被使用梯度下降優化算法來尋找這個函數的最小值點。
- 定義梯度下降算法（gradient_descent）需要兩個參數，即學習率（learning_rate）和迭代次數（num_iterations）。從初始值 x = -5 開始，然後在每個迭代步驟中計算函數的梯度（導數），並根據梯度的方向和學習率來更新 x 的值，並將每個 x 的值記錄在 history 中。
- 設置學習率為 0.1，進行了 50 次迭代，這些值可以根據實際問題進行調整。
- 使用梯度下降算法對函數進行優化，並將優化過程中的 x 值記錄在 trajectory 中。
- 最後繪製函數 f(x) 的曲線以及梯度下降過程中 x 值的軌跡。函數曲線是一個凹向上的二次函數，紅色點表示梯度下降算法在每個步驟中的位置變化。

這個程式碼演示了梯度下降算法在尋找函數最小值時的應用，這是機器學習和優化中的一個重要技巧。通過不斷調整學習率和迭代次數，可以更精確地找到函數的最小值點。

$$f(x) = (x^2+y^2-1)^3=x^2y^3\\教師節快樂$$ 


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a set of x and y values
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)

# Create a grid
X, Y = np.meshgrid(x, y)

# Calculate the left-hand and right-hand sides of the equation
left_side = (X**2 + Y**2 - 1)**3
right_side = X**2 * Y**3

# Plot the left-hand and right-hand sides of the equation
plt.contour(X, Y, left_side - right_side, levels=[0], colors='red')
plt.contourf(X, Y, left_side - right_side, levels=[-1,0,1], colors=['red', 'white'], extend='both')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Happy Teachers Day', fontweight='bold', pad=10)

plt.grid(False)

plt.show()

```

**<font color=skyblue>注意事項與討論：</font>**
- 使用NumPy生成一組x和y的值，範圍從-2到2，總共生成400個點，這將用作此圖的座標。
- 使用np.meshgrid函數來創建一個網格。這將生成X和Y兩個矩陣，其中X和Y的維度相同，包含了所有可能的(x, y)座標組合。
- 使用plt.contour函數繪製方程的等高線圖。在這裡，我們指定left_side - right_side作為等高線圖的函數，然後指定levels=[0]以繪製方程等於0的等高線。這些等高線將以紅色線條顯示。
- 使用plt.contourf函數繪製填充區域。我們指定left_side - right_side作為等高線圖的函數，然後指定levels=[-1, 0, 1]以定義三個不同的水平。這三個水平將使用紅色、白色區域來填充，以突出等高線圖的不同區域。
